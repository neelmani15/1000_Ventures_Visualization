```python
# Importing the libary
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# Generate random value of 1000 ventures from 2000 to 2010
# The value of revenue is between range Initial_amount and Final_amount 
year_wise=[]
Initial_amount=int(input("enter the Initial amount: "))
Final_amount=int(input("Enter the Final amount: "))
for i in range(2000,2011):
    vectors=np.random.randint(Initial_amount,Final_amount,1000)
    year_wise.append(vectors)
```
